```python
# ============================================================
# [2] IMPORT LIBRARIES
# ============================================================

import pandas as pd
import gradio as gr
import math

print("Libraries imported successfully.")
```

```python
# ============================================================
# [3] USER INPUT TRANSACTION DATA
# ============================================================

# Ask the user to enter transaction information

date = input("Enter the transaction date (YYYY-MM-DD): ")

amount = input("Enter the amount (Example: Rs 450): ")

category = input("Enter the category: ")

description = input("Enter the description: ")

# Create dictionary with user input
transaction_data = {
    "Date": [date],
    "Amount": [amount],
    "Category": [category],
    "Description": [description]
}

# Create dataframe
df_sample = pd.DataFrame(transaction_data)

# Show the dataframe
df_sample
```

```python
# ============================================================
# [4] CLEAN AMOUNT FUNCTION
# ============================================================

def clean_amount(amount):
    """
    Clean the amount and convert it into a number.

    Args:
        amount: The money amount entered by the user.

    Returns:
        float: Cleaned amount.
    """

    # Convert amount into text
    amount = str(amount)

    # Check if amount is negative
    is_negative = "-" in amount

    # Remove symbols and spaces
    amount = amount.replace("Rs", "")
    amount = amount.replace("$", "")
    amount = amount.replace(",", "")
    amount = amount.replace("-", "")
    amount = amount.strip()

    # Convert to number
    amount = float(amount)

    # Make negative if needed
    if is_negative:
        amount = -amount

    return amount
```

```python
# ============================================================
# [5] TEST CLEAN AMOUNT FUNCTION
# ============================================================

# Clean the Amount column
df_sample["Amount"] = df_sample["Amount"].apply(clean_amount)

# Show updated dataframe
df_sample
```

```python
# ============================================================
# [6] ANALYSE SPENDING FUNCTION
# ============================================================

def analyze_spending(df):
    """
    Analyse the user's spending data.

    Args:
        df: Transaction dataframe.

    Returns:
        dict: Spending analysis results.
    """

    # Only use positive amounts
    spending_df = df[df["Amount"] > 0]

    # Calculate total spending
    total_spending = spending_df["Amount"].sum()

    # Group spending by category
    category_spending = spending_df.groupby("Category")["Amount"].sum()

    # Sort highest to lowest
    category_spending = category_spending.sort_values(ascending=False)

    # Find highest spending category
    highest_category = category_spending.index[0]

    highest_amount = category_spending.iloc[0]

    # Store results
    results = {
        "total_spending": total_spending,
        "category_spending": category_spending,
        "highest_category": highest_category,
        "highest_amount": highest_amount
    }

    return results
```

```python
# ============================================================
# [7] RUN SPENDING ANALYSIS
# ============================================================

# Run the spending analysis
results = analyze_spending(df_sample)

# Display the results
print("TOTAL SPENDING:")
print("Rs", results["total_spending"])

print("\nSPENDING BY CATEGORY:")
print(results["category_spending"])

print("\nHIGHEST SPENDING CATEGORY:")
print(results["highest_category"])

print("AMOUNT:")
print("Rs", results["highest_amount"])
```

```python
# ============================================================
# [8] GENERATE SAVINGS ADVICE FUNCTION
# ============================================================

def generate_savings_advice(results):
    """
    Generate savings advice based on spending.

    Args:
        results: Spending analysis results.

    Returns:
        str: Savings advice.
    """

    # Get information from results
    total_spending = results["total_spending"]

    highest_category = results["highest_category"]

    highest_amount = results["highest_amount"]

    # Calculate possible savings
    possible_saving = highest_amount * 0.20

    # Create advice message
    advice = f"""
[SAVINGS COACH ADVICE]

Your total spending is Rs {total_spending:.2f}.

Your highest spending category is [{highest_category}].

You spent Rs {highest_amount:.2f} in this category.

If you reduce your [{highest_category}] spending by 20%,
you could save about Rs {possible_saving:.2f}.

Try setting a weekly spending limit for this category.
"""

    return advice
```

```python
# ============================================================
# [9] DISPLAY SAVINGS ADVICE
# ============================================================

# Generate advice
advice = generate_savings_advice(results)

# Show advice
print(advice)
```

```python
# ============================================================
# [10] SAVINGS GOAL CALCULATOR
# ============================================================

def savings_goal_calculator():
    """
    Calculate how long it will take to reach a savings goal.
    """

    # Ask the user for savings information
    current_savings = float(input("Enter your current savings: Rs "))

    monthly_savings = float(input("Enter your monthly savings: Rs "))

    savings_goal = float(input("Enter your savings goal: Rs "))

    # Validation
    if current_savings < 0 or monthly_savings < 0 or savings_goal < 0:
        print("Please enter positive numbers only.")
        return

    # Check if goal is already reached
    if savings_goal <= current_savings:
        print("You have already reached your savings goal.")
        return

    # Check if monthly savings is zero
    if monthly_savings == 0:
        print("Monthly savings must be more than 0.")
        return

    # Calculate amount still needed
    amount_needed = savings_goal - current_savings

    # Calculate months needed
    months_needed = amount_needed / monthly_savings

    # Round up the months
    months_needed = math.ceil(months_needed)

    # Show result
    print("\n[SAVINGS GOAL RESULT]")

    print(f"You need about {months_needed} month(s) to reach your savings goal.")
```

```python
# ============================================================
# [11] RUN SAVINGS GOAL CALCULATOR
# ============================================================

savings_goal_calculator()
```

```python
# ============================================================
# [23] SIMPLE GRADIO APPLICATION
# ============================================================

def savings_coach_app(user_message):
    """
    Give savings advice using a simple Gradio app.

    Args:
        user_message: Message typed by the user.

    Returns:
        str: Finance advice.
    """

    user_message = user_message.lower()

    if "save" in user_message:
        return "Try saving money before spending on unnecessary items."

    elif "budget" in user_message:
        return "A good budget separates money into needs, wants, and savings."

    elif "coffee" in user_message:
        return "Reducing coffee purchases can help improve monthly savings."

    elif "groceries" in user_message or "food" in user_message:
        return "Using a shopping list can reduce unnecessary grocery spending."

    elif "entertainment" in user_message:
        return "Set a monthly entertainment limit to improve savings."

    else:
        return "I am Savings Coach AI. Ask me about saving, budgeting, coffee, groceries, or entertainment."


app = gr.Interface(
    fn=savings_coach_app,
    inputs="text",
    outputs="text",
    title="Savings Coach AI",
    description="A simple Smart Finance Assistant that gives savings and budgeting advice."
)

app.launch()
```
