```python
print("=== Expense Classifier ===")

amount = float(input("Enter an expense amount: $"))
while amount < 0:
    print("Try again. Amount cannot be negative.")
    amount = float(input("Enter an expense amount: $"))

if amount >= 100:
    print("Major Expense - this is a big cost, think carefully.")
elif amount >= 50:
    print("Moderate Expense - noticeable spending, track it.")
elif amount >= 10:
    print("Minor Expense - small but still important.")
else:
    print("Small Purchase - low cost, but small costs add up.")
```

```python
print("=== Budget Limit Checker ===")

monthly_budget = float(input("Enter your monthly budget: $"))
while monthly_budget <= 0:
    print("Try again. Budget must be more than 0.")
    monthly_budget = float(input("Enter your monthly budget: $"))

expense = float(input("Enter expense amount: $"))
while expense < 0:
    print("Try again. Expense cannot be negative.")
    expense = float(input("Enter expense amount: $"))

percent = (expense / monthly_budget) * 100
print(f"\nThis is {percent:.2f}% of your budget.")

if expense > 0.50 * monthly_budget:
    print("Major purchase - think carefully.")
elif expense > 0.25 * monthly_budget:
    print("Significant expense.")
elif expense > 0.10 * monthly_budget:
    print("Moderate expense.")
else:
    print("Minor expense - within budget.")
```

```python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
while goal_amount <= 0:
    print("Try again. Goal must be more than 0.")
    goal_amount = float(input("Savings goal: $"))

current_savings = float(input("Current savings: $"))
while current_savings < 0:
    print("Try again. Savings cannot be negative.")
    current_savings = float(input("Current savings: $"))

percent = (current_savings / goal_amount) * 100
print(f"\nProgress: {percent:.2f}%")

if percent >= 100:
    print("Goal achieved! Consider a new goal.")
elif percent >= 75:
    print("Almost there! You're doing great.")
elif percent >= 50:
    print("Halfway there - keep it up!")
elif percent >= 25:
    print("Good progress, stay consistent.")
else:
    print("Just getting started - you've got this!")
```

```python
print("=== Smart Purchase Advisor ===")

available_money = float(input("How much money do you have available? $"))
while available_money <= 0:
    print("Try again. Available money must be more than 0.")
    available_money = float(input("How much money do you have available? $"))

purchase_price = float(input("How much does the item cost? $"))
while purchase_price < 0:
    print("Try again. Price cannot be negative.")
    purchase_price = float(input("How much does the item cost? $"))

monthly_income = float(input("What's your monthly income? $"))
while monthly_income <= 0:
    print("Try again. Income must be more than 0.")
    monthly_income = float(input("What's your monthly income? $"))

# Basic affordability check
if purchase_price > available_money:
    needed = purchase_price - available_money
    print(f"\nSave up first - you need ${needed:.2f} more.")
else:
    percent_available = (purchase_price / available_money) * 100
    percent_income = (purchase_price / monthly_income) * 100

    print(f"\nThis is {percent_available:.2f}% of your available money.")
    print(f"This is {percent_income:.2f}% of your monthly income.")

    if percent_available >= 50:
        print("You can afford it, but it's expensive. Consider waiting.")
    elif percent_available >= 20:
        print("This seems like a reasonable purchase for your budget.")
    else:
        print("This is a minor expense - go for it!")
```
