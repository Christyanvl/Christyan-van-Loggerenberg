### Step 5: Transform Your Core Functions

Convert your Week 5 functions to work with dictionaries:

```python
# ============================================================================
# ENHANCED EXPENSE MANAGEMENT FUNCTIONS (Fixed - No pyinputplus needed)
# ============================================================================

from datetime import datetime  # make sure this is run in your notebook

def generate_expense_id(tracker):
    tracker["statistics"]["last_expense_id"] += 1
    return f"EXP{tracker['statistics']['last_expense_id']:03d}"

def add_expense_enhanced(tracker, amount, description, category="Other", essential=True):
    # Validation: no negative amounts
    while amount < 0:
        print("Try again. Amount cannot be negative.")
        amount = float(input("Enter amount: "))

    # If user typed a category that does not exist, use "Other"
    if category not in tracker["categories"]:
        category = "Other"
        if "Other" not in tracker["categories"]:
            tracker["categories"]["Other"] = {"budget": 0.0, "spent": 0.0}

    expense_id = generate_expense_id(tracker)
    current_date = datetime.now().strftime("%Y-%m-%d")

    expense = {
        "amount": float(amount),
        "description": description,
        "category": category,
        "date": current_date,
        "essential": essential,
        "added_timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

    tracker["expenses"][expense_id] = expense

    # Update category spending
    tracker["categories"][category]["spent"] += float(amount)

    # Update statistics
    tracker["statistics"]["total_expenses"] += float(amount)
    tracker["statistics"]["expense_count"] += 1

    return expense_id

def find_expense_by_description(tracker, search_term):
    matches = {}
    search_lower = search_term.lower()

    for expense_id, expense_data in tracker["expenses"].items():
        if search_lower in expense_data["description"].lower():
            matches[expense_id] = expense_data

    return matches

def get_expenses_by_category(tracker, category):
    category_expenses = {}

    for expense_id, expense_data in tracker["expenses"].items():
        if expense_data["category"] == category:
            category_expenses[expense_id] = expense_data

    return category_expenses

def calculate_category_summary(tracker):
    summary = {}

    for category, info in tracker["categories"].items():
        budget = float(info.get("budget", 0.0))
        spent = float(info.get("spent", 0.0))
        remaining = budget - spent
        percentage = (spent / budget * 100) if budget > 0 else 0

        summary[category] = {
            "budget": budget,
            "spent": spent,
            "remaining": remaining,
            "percentage": percentage,
            "status": "Over budget" if spent > budget else "Within budget",
            "expense_count": len(get_expenses_by_category(tracker, category))
        }

    return summary

# --------------------------------------------------------------------------
# Test your enhanced functions
# --------------------------------------------------------------------------
print("🧪 Testing enhanced functions:\n")

exp1 = add_expense_enhanced(finance_tracker, 450, "Monthly rent payment", "Housing", True)
exp2 = add_expense_enhanced(finance_tracker, 67.50, "Weekly groceries from Woolworths", "Food", True)
exp3 = add_expense_enhanced(finance_tracker, 89, "Petrol for weekend trip", "Transport", True)

print(f"Added expenses: {exp1}, {exp2}, {exp3}")
print(f"Total expenses: {finance_tracker['statistics']['expense_count']}")
print(f"Total spent: ${finance_tracker['statistics']['total_expenses']:.2f}")
```

### Step 6: Enhanced Analysis Functions

Create more sophisticated analysis using dictionary power:

```python
# ============================================================================
# ENHANCED ANALYSIS FUNCTIONS (Fixed - safe + works without pyinputplus)
# ============================================================================

from datetime import datetime  # make sure this is run

def generate_financial_report(tracker):
    user_info = tracker.get("user_info", {})
    stats = tracker.get("statistics", {})
    monthly_budget = float(user_info.get("monthly_budget", 0.0))

    print("=" * 60)
    print(f"📊 FINANCIAL REPORT FOR {user_info.get('name','').upper()}")
    print("=" * 60)
    print(f"📅 Report Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
    print(f"🗓️  Tracker Created: {user_info.get('tracker_created','')}")
    print()

    total_spent = float(stats.get("total_expenses", 0.0))
    budget_percentage = (total_spent / monthly_budget * 100) if monthly_budget > 0 else 0
    remaining_budget = monthly_budget - total_spent

    print("💰 BUDGET OVERVIEW:")
    print(f"   Monthly Budget: ${monthly_budget:.2f}")
    print(f"   Total Spent: ${total_spent:.2f}")
    print(f"   Remaining: ${remaining_budget:.2f}")
    print(f"   Budget Used: {budget_percentage:.1f}%")
    print()

    print("📊 CATEGORY BREAKDOWN:")
    category_summary = calculate_category_summary(tracker)
    for category, info in category_summary.items():
        status_icon = "⚠️" if info["spent"] > info["budget"] else "✅"
        print(
            f"   {status_icon} {category}: ${info['spent']:.2f} / ${info['budget']:.2f} "
            f"({info['percentage']:.1f}%) - {info['expense_count']} expenses"
        )
    print()

    print("📝 RECENT EXPENSES:")
    expenses_items = list(tracker.get("expenses", {}).items())
    recent_expenses = expenses_items[-5:]

    if len(recent_expenses) == 0:
        print("   No expenses added yet.")
    else:
        for expense_id, expense_data in recent_expenses:
            essential_icon = "🔴" if expense_data.get("essential", True) else "🟡"
            print(
                f"   {expense_id}: ${float(expense_data.get('amount',0)):.2f} - {expense_data.get('description','')} "
                f"({expense_data.get('category','')}) {essential_icon}"
            )
    print()

    print("💡 FINANCIAL INSIGHTS:")
    if budget_percentage >= 90:
        print("   🚨 You're close to your budget limit - consider reducing non-essential spending")
    elif budget_percentage >= 75:
        print("   ⚡ Approaching budget limit - monitor spending carefully")
    elif budget_percentage >= 50:
        print("   📊 Halfway through budget - maintaining good pace")
    else:
        print("   🌟 Excellent budget management - consider increasing savings")

    if tracker.get("expenses", {}):
        largest_expense = max(tracker["expenses"].items(), key=lambda x: float(x[1].get("amount", 0)))
        print(
            f"   💸 Largest expense: {largest_expense[0]} - ${float(largest_expense[1].get('amount',0)):.2f} "
            f"({largest_expense[1].get('description','')})"
        )

    print("=" * 60)


def find_optimization_opportunities(tracker):
    opportunities = []
    category_summary = calculate_category_summary(tracker)

    for category, info in category_summary.items():
        if info["spent"] > info["budget"]:
            overspend = info["spent"] - info["budget"]
            opportunities.append({
                "type": "overspend",
                "category": category,
                "amount": overspend,
                "message": f"Reduce {category} spending by ${overspend:.2f} to stay within budget"
            })

    for category, info in category_summary.items():
        if info["percentage"] < 25 and info["expense_count"] == 0:
            opportunities.append({
                "type": "underutilized",
                "category": category,
                "amount": info["budget"],
                "message": f"Consider reallocating ${info['budget']:.2f} from unused {category} budget"
            })

    monthly_budget = float(tracker.get("user_info", {}).get("monthly_budget", 0.0))
    expensive_threshold = monthly_budget * 0.05

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        amount = float(expense_data.get("amount", 0.0))
        essential = expense_data.get("essential", True)

        if amount >= expensive_threshold and not essential:
            opportunities.append({
                "type": "expensive_item",
                "expense_id": expense_id,
                "amount": amount,
                "message": f"Review {expense_id}: ${amount:.2f} for {expense_data.get('description','')}"
            })

    return opportunities


# 🧪 Test the enhanced reporting
print("🧪 Testing enhanced analysis:\n")

generate_financial_report(finance_tracker)

opportunities = find_optimization_opportunities(finance_tracker)
if opportunities:
    print("\n🎯 OPTIMIZATION OPPORTUNITIES:")
    for i, opp in enumerate(opportunities, 1):
        print(f"   {i}. {opp['message']}")
else:
    print("\n✅ No optimization opportunities found - excellent financial management!")
```

## 🎯 Phase 3: Progressive Enhancement (30 minutes)

### Step 7: Add Advanced Dictionary Features

Implement sophisticated features that would be difficult with lists:

```python
# ============================================================================
# ADVANCED DICTIONARY FEATURES (Fixed - Safe Version)
# ============================================================================

from datetime import datetime  # make sure this is run

def create_expense_filter(tracker, **criteria):
    filtered_expenses = {}

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        include_expense = True

        for criterion, value in criteria.items():
            if criterion == "category" and expense_data.get("category") != value:
                include_expense = False
                break
            elif criterion == "essential" and expense_data.get("essential") != value:
                include_expense = False
                break
            elif criterion == "min_amount" and float(expense_data.get("amount", 0)) < value:
                include_expense = False
                break
            elif criterion == "max_amount" and float(expense_data.get("amount", float("inf"))) > value:
                include_expense = False
                break
            elif criterion == "date" and expense_data.get("date") != value:
                include_expense = False
                break
            elif criterion == "description_contains":
                if value.lower() not in expense_data.get("description", "").lower():
                    include_expense = False
                    break

        if include_expense:
            filtered_expenses[expense_id] = expense_data

    return filtered_expenses


def create_monthly_summary_by_date(tracker):
    monthly_summary = {}

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        expense_date = expense_data.get("date", "")
        if expense_date:
            year_month = expense_date[:7]

            if year_month not in monthly_summary:
                monthly_summary[year_month] = {
                    "total_amount": 0.0,
                    "expense_count": 0,
                    "categories": {},
                    "expenses": {}
                }

            amount = float(expense_data.get("amount", 0.0))

            monthly_summary[year_month]["total_amount"] += amount
            monthly_summary[year_month]["expense_count"] += 1
            monthly_summary[year_month]["expenses"][expense_id] = expense_data

            category = expense_data.get("category", "Other")
            if category not in monthly_summary[year_month]["categories"]:
                monthly_summary[year_month]["categories"][category] = 0.0
            monthly_summary[year_month]["categories"][category] += amount

    return monthly_summary


def export_expense_data(tracker, format="summary"):
    if format == "summary":
        return {
            "export_date": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "user": tracker.get("user_info", {}).get("name", ""),
            "total_budget": float(tracker.get("user_info", {}).get("monthly_budget", 0.0)),
            "total_spent": float(tracker.get("statistics", {}).get("total_expenses", 0.0)),
            "total_expenses": int(tracker.get("statistics", {}).get("expense_count", 0)),
            "category_summary": calculate_category_summary(tracker)
        }

    elif format == "detailed":
        return {
            "export_date": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "complete_tracker": tracker
        }

    elif format == "csv_ready":
        csv_data = []
        for expense_id, expense_data in tracker.get("expenses", {}).items():
            csv_data.append({
                "expense_id": expense_id,
                "amount": float(expense_data.get("amount", 0.0)),
                "description": expense_data.get("description", ""),
                "category": expense_data.get("category", "Other"),
                "date": expense_data.get("date", ""),
                "essential": expense_data.get("essential", True)
            })
        return csv_data


# --------------------------------------------------------------------------
# Test advanced features
# --------------------------------------------------------------------------
print("🧪 Testing advanced dictionary features:\n")

# Make sure these categories exist before adding test expenses
for cat in ["Entertainment", "Utilities", "Food", "Other"]:
    if cat not in finance_tracker["categories"]:
        finance_tracker["categories"][cat] = {"budget": 0.0, "spent": 0.0}

add_expense_enhanced(finance_tracker, 25, "Coffee with friends", "Entertainment", False)
add_expense_enhanced(finance_tracker, 150, "Electric bill", "Utilities", True)
add_expense_enhanced(finance_tracker, 45, "Lunch at restaurant", "Food", False)

print("🔍 FILTERING EXAMPLES:")

food_expenses = create_expense_filter(finance_tracker, category="Food")
print(f"Food expenses: {len(food_expenses)} found")

expensive_expenses = create_expense_filter(finance_tracker, min_amount=100)
print(f"Expensive expenses (>$100): {len(expensive_expenses)} found")

non_essential = create_expense_filter(finance_tracker, essential=False)
print(f"Non-essential expenses: {len(non_essential)} found")

monthly_data = create_monthly_summary_by_date(finance_tracker)
print("\n📅 MONTHLY SUMMARY:")
if monthly_data:
    for month, data in monthly_data.items():
        print(f"   {month}: ${data['total_amount']:.2f} ({data['expense_count']} expenses)")
else:
    print("   No monthly data yet.")

export_summary = export_expense_data(finance_tracker, "summary")
print("\n📤 EXPORT READY:")
print(f"   Total Budget: ${export_summary['total_budget']:.2f}")
print(f"   Total Spent: ${export_summary['total_spent']:.2f}")
print(f"   Export contains: {len(export_summary)} main sections")
```
