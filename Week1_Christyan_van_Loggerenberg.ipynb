## Part 3: Building the Budget Tracker

Use AI to help you write a program that:

- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
1) Ask the user for their name.
2) Ask for monthly income in Mauritian Rupees (MUR).
3) Ask for 3 expenses in MUR.
4) Calculate total expenses and remaining budget.
5) Print a summary with MUR values shown to 2 decimal points.
6) If the remaining budget is negative, print a warning message.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
name = input("Enter your name: ")
income = float(input("Enter your monthly income in MUR: "))

expense1 = float(input("Enter expense 1 in MUR: "))
expense2 = float(input("Enter expense 2 in MUR: "))
expense3 = float(input("Enter expense 3 in MUR: "))

total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

print("\n----- BUDGET SUMMARY -----")
print("Name:", name)
print(f"Income: MUR {income:.2f}")
print(f"Total Expenses: MUR {total_expenses:.2f}")
print(f"Remaining Budget: MUR {remaining:.2f}")

if remaining < 0:
    print("⚠️ WARNING: Your budget is negative!")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: Input validation (no negative numbers)"
enhancement_prompt = """
I asked the AI to add input validation so the user cannot enter negative numbers for income or expenses.
If a negative number is entered, the program should ask again until a valid number is entered.
Keep the program simple and beginner-friendly.
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
name = input("Enter your name: ")

income = float(input("Enter your monthly income in MUR: "))
while income < 0:
    print("Income cannot be negative. Please try again.")
    income = float(input("Enter your monthly income in MUR: "))

expense1 = float(input("Enter expense 1 in MUR: "))
while expense1 < 0:
    print("Expense cannot be negative. Please try again.")
    expense1 = float(input("Enter expense 1 in MUR: "))

expense2 = float(input("Enter expense 2 in MUR: "))
while expense2 < 0:
    print("Expense cannot be negative. Please try again.")
    expense2 = float(input("Enter expense 2 in MUR: "))

expense3 = float(input("Enter expense 3 in MUR: "))
while expense3 < 0:
    print("Expense cannot be negative. Please try again.")
    expense3 = float(input("Enter expense 3 in MUR: "))

total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

print("\n----- BUDGET SUMMARY -----")
print("Name:", name)
print(f"Income: MUR {income:.2f}")
print(f"Total Expenses: MUR {total_expenses:.2f}")
print(f"Remaining Budget: MUR {remaining:.2f}")

if remaining < 0:
    print("⚠️ WARNING: Your budget is negative! You spent more than you earn.")
else:
    print("✅ You are within your budget.")
```
